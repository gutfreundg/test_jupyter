```python
def convert_week(year_and_week_int):
    year_and_week_str=str(year_and_week_int)
    year=int(year_and_week_str[:4])
    week=int(year_and_week_str[4:])
    w=isoweek.Week(year,week)
    return pd.Period(w.day(0),'W')
```
